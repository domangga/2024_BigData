```python
import pandas as pd
import numpy as np
exam = pd.read_csv('exam.csv')
mpg = pd.read_csv('mpg.csv')
```

```python
exam.agg(mean_math=('math','mean'))
```

## 집단별 요약 통계향 구하기

```python
# nclass별로 분리하기.groupby()
# math 평균 .agg()
exam.groupby('nclass').agg(math=('math','mean'),english=('english','mean'),science=('science','mean'))
```

```python
#nclass별로분ㄹ
#수학점수 평균 mean
# 수학점수 합계 sum
#수학점수 중앙값 median
# 빈도(학생수)count
exam.groupby('nclass').agg(mean_math=('math','mean'),sum_math=('math','sum'),median_math=('math','median'), n=('nclass','count'))
```

```python
mpg.query('manufacturer == "audi"').groupby('drv').agg(n=('drv','count'))
```

```python
# chevrolet의 drv별 빈도
#chevrolet 추출
#drv뱔 분리
# 빈도 구하기
mpg.groupby('drv').agg(n=('drv','count'))
```
