```python
import pandas as pd

df=pd.DataFrame({'var1':[4,3,8],'var2':[2,6,1]})

df
```

```python
#var_sum 열추가

df['var_sum']=df['var1']+df['var2']
```

```python
#var_mean 파생변수 만들기

df['var_mean']=df['var_sum']/2
```

```python
mpg=pd.read_csv('mpg.csv')
```

```python
mpg.head()
```

```python
mpg['total']=(mpg['cty']+mpg['hwy'])/2
```

```python
mpg
```

```python
#평균구하는 방법

#sum/n
#.mean()
print(mpg['total'].sum()/len(mpg))
print(mpg['total'].mean())
```

```python
mpg['total'].plot.hist()#빈도수
```

```python
import numpy as np
```

```python
#20 이상이면 pass 그렇지 않으면 fail 부여한다
mpg['test']=np.where(mpg['total']>=20,'pass','fail')
```

# 빈도표로 합격 판정 자돋차 수 살펴보기

```python
#연비 합격 숫자
mpg['test'].value_counts()
```

```python
# 박대 그래프로 빈도 표현하기
count_test = mpg['test'].value_counts()
```

```python
count_test.plot.bar(rot=0)#축ㅇ 이름 수평 설정
```

```python
# total 기준으로 A,B,C등급 부여
mpg['grade']=np.where(mpg['total']>=30,'A',np.where(mpg['total']>=20,'B','C'))
```

```python
mpg.head()
```

```python
# A,B,C 빈도표 만들기 
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```

```python
# 알파벳순 정렬
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```
