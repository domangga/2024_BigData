## 원형 차트 그리기

```python
plt.rc('font',family='Malgun Gothic')

#data 혈액형 수
b_type=[25,19,37,11]
b_name=["A형",'B형','O형','AB형']
#제목:혈액형 분포
plt.title('혈액형 분포')
#원형 차트 그리는 명령어 pie 범례 labels=[] autopct=''
plt.pie(b_type, labels=b_name,autopct='%1.1f%%')
plt.show()
```

```python

kor=[80,20,50,20,10,50,60,30,40]
eng=[90,40,10,40,60,20,50,70,90]
#산점도 글기 scatter option color = red 투명도 alpha 값
plt.scatter(kor,eng, color='r',alpha=0.3)
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
```
