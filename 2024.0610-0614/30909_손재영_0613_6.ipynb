```python
import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic')

sales=[1,5,7,3,7]
month=[3,4,5,6,7]

plt.title('월별 판매 실적')
plt.bar(month,sales,color='b')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

sales =[1,5,7,3,7]
month=[3,4,5,6,7]

# 제목
plt.title('월별 판매 실적')

#가로 바차트 그리기 barh range(시작,마지막+1)
plt.barh(range(1,5+1),sales,color='red')
plt.show()
```
